```python
import pickle
import csv
```

```python
with open('showscore_scores_descriptions.csv', mode='r') as infile:
    reader = csv.reader(infile)

    mydict = {rows[0]:rows[1:3] for rows in reader}
```

```python
showscore = {}
for show in mydict:
    showscore[show] = {}
    showscore[show]['score'] = mydict[show][0]
    showscore[show]['description'] = mydict[show][1]
```
