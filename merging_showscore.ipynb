```python
import pickle
import csv
```

```python
with open('showscore_scores_descriptions.csv', mode='r') as infile:
    reader = csv.reader(infile)

    mydict = {rows[0]:rows[1:3] for rows in reader}
```

```python
showscore = {}
for show in mydict:
    showscore[show] = {}
    showscore[show]['score'] = mydict[show][0]
    showscore[show]['description'] = mydict[show][1]
```

```python
for s in showscore.keys():
    print(s)
```

```python
for s in showscore:
    if "&" in s and s not in ["Song &amp; Dance", "Mack &amp; Mabel", "Jekyll &amp; Hyde"]:
        new_key = s[:s.index("&amp;")] + "and" + s[s.index("&amp;")+5:]
        showscore[new_key] = showscore.pop(s)
```

```python
new_showscore = dict(showscore)
```
